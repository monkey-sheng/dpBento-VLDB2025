```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme('paper', font_scale=1.35, style='whitegrid', rc={"axes.titlesize": 18, "axes.labelsize": 16,})
```

```python
# ml micro benchmark
# fp32 results

x_name = '# Threads'
# x_val = ['B0-fp32', 'B0-int8']
x_val = [1,2,4,8,16]  # num_threads
y_name = 'inference latency (ms)'

var_name = 'Platform'

bf3_vals = np.array([35.15265088,19.92067052,12.54798276,8.79160308,7.8535418])
bf2_vals = np.array([84.34220768,49.22381112,32.46268134,25.449489,288.0070945])
host_vals = np.array([9.48382612,8.28611118,6.42626768,5.07056314,5.82787908])

# BF-2 using dub data
raw_data = pd.DataFrame.from_dict({x_name: x_val, 'BF-2': bf2_vals, 'BF-3': bf3_vals, 'Host CPU': host_vals})
data = pd.melt(raw_data, id_vars=x_name, value_name=y_name, var_name=var_name)

plt.figure(figsize=(8, 4))
ax = sns.barplot(data=data, x=x_name, y=y_name, hue=var_name, gap=0.12)
# display the value on top of the bar
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(3, 10), textcoords='offset points', size=12, weight='bold')
# ax.set_yscale("log")
ax.set_xticklabels(ax.get_xticklabels(), size=18)
ax.set_yticklabels(ax.get_yticklabels(), size=18, weight='bold')
ax.set_title('Image Classification - fp32', size=20, weight='bold', pad=20)

plt.legend(fontsize=13.8, loc='upper left',)
plt.tight_layout()
# plt.show()

plt.savefig('ml-480p-fp32-diff_threads.pdf')
```
